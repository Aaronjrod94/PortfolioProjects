```python
#Importing relevant Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# Creating Data Frame

df = pd.read_csv('movies.csv')
df.head()
```

```python
# Seeing how many rows

df.info()
```

```python
# Looking at data

df.head()
```

```python
#Checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for columns

df.dtypes
```

```python
#Changing data type (Float64 to int64) for better readability
df.fillna({'budget': 0}, inplace=True)
df['budget'] = df['budget'].astype('int64')
```

```python
df.fillna({'gross': 0}, inplace=True)
df['gross'] = df['gross'].astype('int64')
```

```python
# Sorting films by gross in descending order

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Setting up the option to display every film

pd.set_option('display.max_rows',None)
```

```python
df.sort_values(by=['gross'], inplace = True, ascending = False)
df.head()
```

```python
# Drop any duplicates

df.drop_duplicates()
df.head()
```

```python
# Budget will possibly have a high correlation
```

```python
#Scatter plot with budget vs gross
```

```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.show()
```

```python
#High correlation between budget and gross
```

```python
#Better visualization for Correlations 

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#Looking at company

```

```python
df_numbered = df

for col_name in df_numbered.columns:
    if(df_numbered[col_name].dtype == 'object'):
        df_numbered[col_name] = df_numbered[col_name].astype('category')
        df_numbered[col_name] = df_numbered[col_name].cat.codes
df_numbered
df_numbered.head()
            
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df.head()
```

```python
correlation_matrix = df_numbered.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#Unstacking
correlation_mat = df_numbered.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
